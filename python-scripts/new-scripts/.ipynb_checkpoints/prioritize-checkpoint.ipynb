```python
def get_age_map(df):
    numbers = list(df['number'].values) 
    date_created = list(df['created_at'])
    age_list = zip(numbers, date_created)
    sorted_age_list = sorted(age_list, key = lambda tuple: tuple[1])
    age_map = {}
    i = 1
    length = len(sorted_age_list)
    for tup in sorted_age_list:
        age_map[tup[0]] = i / length
        i += 1
    return age_map    
```

```python
# members and contributors have been given equal weightage 
def get_contributor_map(df):
    numbers = list(df['number'].values)
    author_assoc = list(df['author_association'].values)
    author_assoc1 = []
    for author in author_assoc:
        if(author == 'CONTRIBUTOR' or author == 'MEMBER'):
            author_assoc1 += [0]
        else:
            author_assoc1 += [1]
    author_assoc_list = zip(numbers, author_assoc1)
    sorted_author_assoc = sorted(author_assoc_list, key = lambda tp: tp[1])
    contributor_map = {}
    length = len(sorted_author_assoc)
    for tp in sorted_author_assoc:
        if(tp[1] == 0):
            contributor_map[tp[0]] = 0
        else:
            contributor_map[tp[0]] = 1
    return contributor_map    
```

```python
get_contributor_map(df)
```

```python
class Issue:
    def __init__(self, number, title, comments, assignees, created_at, author_assoc):
        self.comments = comments
        self.assignees = assignees
        self.created_at = created_at
        if(author_assoc == 'MEMBER'):
            self.author_assoc = 1
        elif(author_assoc == 'CONTRIBUTOR'):
            self.author_assoc = 2
        else:
            self.author_assoc = 3
        self.title = title
        self.number = number
```

```python
issue = Issue(1000, 'This is an issue', 3, 5, '1-2-2015', 'MEMBER')
```

```python
import numpy as np
def get_issue_list(df):
    length = df.shape[0]
    issue_list = np.array([])
    for i in range(length):
        series = df.iloc[i,:]
        title = series['title']
        comments = series['comments']
        assignees = series['assignees']
        created_at = series['created_at']
        author = series['author_association']
        number = series['number']
        issue = Issue(number, title, comments, assignees, created_at, author)
        issue_list = np.append(issue_list, issue)
    return issue_list    
```

```python
issue_list = get_issue_list(df)
```

```python
from operator import attrgetter, itemgetter
sorted_list = sorted(issue_list, key = attrgetter('author_assoc'))
sort(key = attrgetter('created_at'))
```

```python
for issue in sorted_list:
    print(issue.number)
```
