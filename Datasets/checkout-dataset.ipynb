```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('tensorflow-issues-closed1.csv', encoding = 'latin-1')
```

```python

date1 = '2017-11-16T16:05:16Z'
date2 = '2017-12-05T18:07:27Z'
import re
def getDays(date1, date2):
    vals1 = re.findall(r'[\d]+', date1)
    vals2 = re.findall(r'[\d]+', date2)
    nums1 = [int(x) for x in vals1]
    nums2 = [int(x) for x in vals2]
    days = (nums2[0] - nums1[0])*365 + (nums2[1] - nums1[1]) * 30 + (nums2[2] - nums1[2])
    return days
```

```python
created_dates = df['created_at'].values
closed_dates = df['closed_at'].values
```

```python
closing_ages = [getDays(date1, date2) for date1, date2 in zip(created_dates, closed_dates)]
```

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(closing_ages,20)
plt.xlabel('days')
plt.ylabel('number of issues')
plt.title('time taken to close issues')
```

```python
author_vs_closed = df[['author_association', 'created_at', 'closed_at']].set_index('author_association')
```

```python
none_created = author_vs_closed.loc['NONE']['created_at'].values
none_closed = author_vs_closed.loc['NONE']['closed_at'].values
mem_created = author_vs_closed.loc['MEMBER']['created_at'].values
mem_closed = author_vs_closed.loc['MEMBER']['closed_at'].values
cont_created = author_vs_closed.loc['CONTRIBUTOR']['created_at'].values
cont_closed = author_vs_closed.loc['CONTRIBUTOR']['closed_at'].values
```

```python
none_age = [getDays(date1, date2) for date1, date2 in zip(none_created, none_closed)]
mem_age = [getDays(date1, date2) for date1, date2 in zip(mem_created, mem_closed)]
cont_age = [getDays(date1, date2) for date1, date2 in zip(cont_created, cont_closed)]
```

```python
f = plt.figure()
f.set_figheight(5)
f.set_figwidth(10)
plt.subplot(131)
plt.hist(none_age,10)
plt.xlabel('days')
plt.ylabel('Number of issues')
plt.title('NONE')
plt.subplot(132)
plt.hist(mem_age, 10)
plt.xlabel('days')
plt.ylabel('number of issues')
plt.title('MEMBER')
plt.subplot(133)
plt.hist(cont_age,10)
plt.xlabel('days')
plt.ylabel('no of issues')
plt.title('CONTRIBUTOR')
```
